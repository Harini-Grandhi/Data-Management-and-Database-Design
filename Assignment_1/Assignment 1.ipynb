```python
#initializing lists to store the collected variables
first_name=[]
middle_name=[]
last_name=[]
DOB=[]
gender=[]
party=[]
twitter_account=[]
in_office=[]
total_votes=[]
missed_votes=[]
state=[]
title=[]
govtrack_id = []
```

```python
#loading data into the list
for each in json_loads2['results'][0]['members']:
        first_name.append(each['first_name'])
        middle_name.append(each['middle_name'])
        last_name.append(each['last_name'])
        DOB.append(each['date_of_birth'])
        gender.append(each['gender'])
        party.append(each['party'])
        twitter_account.append(each['twitter_account'])
        in_office.append(each['in_office'])
        total_votes.append(each['total_votes'])
        missed_votes.append(each['missed_votes'])
        state.append(each['state'])
        title.append(each['title'])
        govtrack_id.append(each['govtrack_id'])
```

```python
#converting lists to dataframe
house_members_df = pd.DataFrame(np.column_stack([first_name,middle_name,last_name,title,state,party,DOB,gender,govtrack_id,total_votes,missed_votes,twitter_account]), 
                               columns=['first_name','middle_name','last_name','title','state','party','DOB','gender','govtrack_id','total_votes','missed_votes','twitter_account'])
```
